````python
def swap_ages(name1, age1, name2, age2):

    age1, age2 = age2, age1
    return name1, age1, name2, age2

name1 = input("Enter the first name: ")
age1 = int(input("Enter the {name1}'s age: "))

name2 = input("Enter the second name: ")
age2 = int(input("Enter the {name2}'s age: "))

name1, age1, name2, age2 = swap_ages(name1, age1, name2, age2)

print(f"After swapping, {name1}'s age is {age1} and {name2}'s age is now {age2}.")
           
````
